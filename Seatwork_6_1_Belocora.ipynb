```python
#Adding a new column named Review Checker which shows if the products has a review
data['Review Checker'] = np.where(data['reviews'] == 0, 'No Reviews', 'With Reviews')
data
```

```python
#Creating a new Dataframe with important datas
basic_data = pd.DataFrame(data, columns=['product_name', 'product_id',
                                        'sale_price', 'rating', 'reviews'])
basic_data
```

```python
#Creating a new Column named Review Checker that shows if the shoes has a Review
basic_data['Review Checker'] = np.where(basic_data['reviews'] == 0,
                                      'No Reviews', 'With Reviews')
basic_data
```

```python
#Prints all the data that has a Review
with_reviews = basic_data.loc[basic_data['Review Checker'] == 'With Reviews']
with_reviews
```

```python
#Prints all the data that has no Reviews
without_reviews = basic_data.loc[basic_data['Review Checker'] == 'No Reviews']
without_reviews
```

```python
#Prints the mean of the sale price of the shoe models
np.mean(basic_data['sale_price'])
```

```python
# Shows the total number of shoes that has a review
num_of_with = len(basic_data[basic_data['Review Checker'] == "With Reviews"])

print('Number of Shoes that has Reviews:', num_of_with)
```

```python
# Shows the total number of shoes that has no review
num_of_without = len(basic_data[basic_data['Review Checker'] == "No Reviews"])

print('Number of Shoes that has no Reviews:', num_of_without)
```

```python
# Shows the shoe with the Highest Price
max_sale_price_index = np.argmax(basic_data['sale_price'])
max_sale_price_row = basic_data.loc[max_sale_price_index]

print("Nike Shoe with the Highest Sale Price:")
print(max_sale_price_row)
```

```python
# Shows the Shoe with the Lowest Price
min_sale_price_index = np.argmin(basic_data['sale_price'])
min_sale_price_row = basic_data.loc[min_sale_price_index]

print("\nNike Shoe with the Lowest Sale Price:")
print(min_sale_price_row)
```

```python
# Shows the shoe with the Highest Ratings
max_rating_index = np.argmax(basic_data['rating'])
max_rating_row = basic_data.loc[max_rating_index]

print("Nike Shoe with the Highest Rating:")
print(max_rating_row)
```

```python
# Shows the shoe with the Lowest Rating
min_rating_index = np.argmin(basic_data['rating'])
min_rating_row = basic_data.loc[min_rating_index]

print("Nike Shoe with the Lowest Rating:")
print(min_rating_row)
```

```python
# Shows the mean of the sale_price, rating, and reviews
mean_sale_price = basic_data['sale_price'].mean()
mean_rating = basic_data['rating'].mean()
mean_reviews = basic_data['reviews'].mean()

print(f"Average of Sale Price: {mean_sale_price}")
print(f"Average of Rating: {mean_rating}")
print(f"Average of Reviews: {mean_reviews}")
```

```python
# Shows the median of the sale_price, rating, and reviews
median_sale_price = basic_data['sale_price'].median()
median_rating = basic_data['rating'].median()
median_reviews = basic_data['reviews'].median()

print(f"Median of Sale Price: {median_sale_price}")
print(f"Median of Rating: {median_rating}")
print(f"Median of Reviews: {median_reviews}")
```

```python
# Shows the mode of the sale_price, rating, and reviews
mode_sale_price = basic_data['sale_price'].mode()
mode_rating = basic_data['rating'].mode()
mode_reviews = basic_data['reviews'].mode()

print(f"Mode of Sale Price: {mode_sale_price}")
print(f"Mode of Rating: {mode_rating}")
print(f"Mode of Reviews: {mode_reviews}")
```
